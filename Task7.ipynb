**1: Smart Light Controller**

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def show_status(self):
        print(self.name, "is", self.status)


# Sample usage
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")

light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
else:
    light.turn_off()

light.show_status()
```

**2: Employee ID Card System**

```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.department = dept

    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample usage
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

emp = Employee(name, emp_id, dept)
emp.display_id_card()
```

**3: Mobile Contact Record**

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def save_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample usage
contact_name = input("Contact Name: ")
phone_number = input("Phone Number: ")

person = Contact(contact_name, phone_number)
person.save_contact()
```

**4: Product Price Tag Generator**

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print("\nProduct:", self.name)
        print("Price: ₹" + str(self.price))


# Sample usage
product_name = input("Product Name: ")
price = int(input("Price: "))

item = Product(product_name, price)
item.print_price_tag()
```

** 5: Movie Rating Display System**

```python
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def show_details(self):
        print("\nMovie:", self.title)
        print("Rating:", self.rating, "/ 5")


# Sample usage
movie_name = input("Movie Name: ")
rating = float(input("Rating: "))

film = Movie(movie_name, rating)
film.show_details()
```

**6: Delivery Address Manager**

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def print_details(self):
        print("\nDelivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


# Sample usage
customer_name = input("Customer Name: ")
address = input("Address: ")

order = Delivery(customer_name, address)
order.print_details()
```
