**1. Social Media – Post Engagement Analyzer**

```python
def analyze_engagement(likes):
    total = 0
    for like in likes:
        total += like

    status = "Viral Post" if total >= 1000 else "Normal Engagement"
    return total, status


likes = [200, 300, 250, 300]
total, status = analyze_engagement(likes)

print("Total Likes:", total)
print("Post Status:", status)
```

**2: Healthcare – Medicine Stock Alert System**

```python
def check_medicine_stock(stock):
    if stock < 10:
        return "Low Stock Alert"
    else:
        return "Stock Sufficient"


stock = 6
print("Medicine Stock:", stock)
print("Status:", check_medicine_stock(stock))
```

**3. Agriculture – Rainfall Adequacy Checker**

```python
def check_rainfall(data, required):
    total = 0
    for rain in data:
        total += rain

    avg = total / len(data)
    status = "Adequate Rainfall" if avg >= required else "Inadequate Rainfall"
    return avg, status


rainfall = [70, 75, 68, 75]
avg, status = check_rainfall(rainfall, 70)

print("Average Rainfall:", avg)
print("Rainfall Status:", status)
```

**4. Social Media – Duplicate Account Detection**

```python
def check_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        return "Yes"
    else:
        return "No"


users = ["ram", "sita", "ram"]
print("Duplicate Accounts Found:", check_duplicates(users))
```

**6. Agriculture – Premium Crop Price Filter**

```python
def filter_premium(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)
    return premium


prices = [1500, 2500, 1800, 3200]
print("Premium Crops:", filter_premium(prices))
```

**7. System Monitoring – Application Health Checker**

```python
def check_health(errors):
    if errors == 0:
        return "Healthy"
    elif errors <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"


errors = 7
print("Error Count:", errors)
print("System Status:", check_health(errors))
```

**8. Banking – Daily Transaction Limit Checker**

```python
def check_transaction(amount):
    limit = 50000
    return "Approved" if amount <= limit else "Rejected"


amount = 60000
print("Transaction Amount:", amount)
print("Transaction Status:", check_transaction(amount))
```

** 9. E-Learning – Student Attendance Eligibility System**

```python
def check_attendance(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    return percentage, status


attendance = ["P", "P", "A", "P", "P"]
percent, status = check_attendance(attendance)

print("Attendance Percentage:", percent)
print("Exam Eligibility:", status)
```

**10. Smart Electricity Bill Analyzer**

```python
def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


units = 250
bill, usage = calculate_bill(units)

print("Total Bill:", bill)
print("Usage Status:", usage)
```
