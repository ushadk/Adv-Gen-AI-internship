```python
#1 Employee Performance Bonus Eligibility


employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Finding the highest score
highest_score = max(employees.values())

#Finding all employees who have this highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)


print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
#2 Search Query Keyword Analysis

# Given query
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

#Removing punctuation
import string
for char in string.punctuation:
    query = query.replace(char, "")

#Split sentence into words
words = query.split()

#Count frequency of each word
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

#Display only words searched more than once
result = {}

for word, count in word_count.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
#3 Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Loop through readings with index
for index in range(len(sensor_readings)):
    value = sensor_readings[index]

    # Check if reading is even (valid)
    if value % 2 == 0:
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#4 Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

#Extract domain and count occurrences
for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

#Calculate percentage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
#5 Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

#Calculate average sales
average_sales = sum(sales) / len(sales)

#Calculate threshold (30% above average)
threshold = average_sales * 1.30

#Detect spike days
for day in range(len(sales)):
    if sales[day] > threshold:
        print(f"Day {day + 1}: {sales[day]}")
```

```python
#6 Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

#Count occurrences of each user ID
for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

#Display duplicates only
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
