**1. Mobile Recharge Validation System**

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        # Rule 1: Minimum recharge
        if amount < 50:
            print("Recharge must be at least ₹50. Try again.\n")
            continue

        # Rule 2: Valid plans
        if amount not in valid_plans:
            print("Invalid plan selected. Available plans:", valid_plans)
            print("Please try again.\n")
            continue

        # If all conditions satisfied
        print("Recharge Successful ✅")
        break

validate_recharge()
```

**2. Inventory Reorder Alert System**

```python
def check_inventory(products):
  for item, quantity in products.items():
    if quantity < 15:
      print(item, "Reorder Alert!")
    else:
      print(item, "Stock OK")

#product stock data
inventory ={
     "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 30
}

check_inventory(inventory)
```

**3. Student Result Processing System**

```python
def calculate_result(marks):
  sum=0
  for mark in marks:
    sum+=mark

  average=sum/len(marks)
  if average >=50:
    print("Pass")
  else:
    print("Fail")

#marks list
marks=list(map(int,input("Enter marks separated by space: ").split()))

result=calculate_result(marks)
```

**4. Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance,peak_hours):
  base_fare=50
  per_km=12

  fare=base_fare+(distance*per_km)
  #for peak hours
  if peak_hours.lower()=="yes":
    fare+=fare*0.25

  return fare
while True:
  distance=float(input("Enter distance in km: "))
  peak_hours=input("Is it peak hours?(yes/no): ")

  total_fare=calculate_fare(distance,peak_hours)
  print("Total Fare: ₹", round(total_fare, 2))

  retry = input("Do you want to calculate again? (yes/no): ")
  if retry.lower() != "yes":
    print("Thank you for using cab service ")
    break
```

**5. Employee Attendance Eligibility**

```python
def check_eligibility(attendance):
  present=0
  for day in attendance:
    if day == "p":
      present+=1

  percentage=(present/len(attendance))*100

  if percentage >=75:
    return percentage, "Eligible"
  else:
    return percentage, "Not Eligible"

#user input
attendance=input("Enter attendance p/a separated by space:").split()

#call function
percent,status=check_eligibility(attendance)
print("Attendance percentage: ", round(percent,2))
print("Status: ",status)
```

**6. Password Strength Checker**

```python
def check_password(password):
  if len(password)<8:
    return "weak Password (Minimum 8 characters required)"\

  has_digit=False
  has_special=False
  special_chars="!@#$%^&*()_+"
  #check for each letter
  for ch in password:
    if ch.isdigit():
      has_digit=True
    if ch in special_chars:
      has_special=True

  #final validation
  if has_digit and has_special:
    return "Strong Password"
  else:
    return "Weak Password((Must contain digit and special character))"

#user input
password=input("Enter password: ")
result=check_password(password)
print(result)
```
