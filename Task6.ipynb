```python
#1 Smart Parking Lot Management System
def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        peak = max(peak, parked)

    status = "Full" if parked >= capacity else "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak)
    print("Parking Status:", status)


parking_system(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
#2 Online Food Delivery Time Estimator
def delivery_eta(distance, traffic, weather):
    eta = distance * 5   # base 5 min per km

    if traffic == "High":
        eta += 15
    elif traffic == "Medium":
        eta += 8

    if weather == "Rainy":
        eta += 10
    elif weather == "Storm":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")


delivery_eta(8, "High", "Rainy")
```

```python
#3 Movie Theatre Seat Occupancy Analyzer
def theatre_analysis(total_seats, booked):
    occupancy = (len(booked) / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    print(f"Occupancy: {occupancy:.0f}%")
    print("Show Status:", status)


theatre_analysis(200, [1]*150)
```

```python
#4 Cloud Server Load Classification
def server_load(readings):
    avg = sum(readings) / len(readings)

    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {avg:.0f}%")
    print("Server Status:", status)


server_load([45, 60, 70, 85, 90])
```

```python
#5 Smart Classroom Resource Monitor
def classroom_usage(resources):
    overused = [k for k, v in resources.items() if v > 8]

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", "Yes" if overused else "No")


classroom_usage({"Projector":6, "AC":9, "Lights":4})
```

```python
#6 Event Registration Capacity Controller
def registration_system(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlist = max(0, registrations - capacity)

    status = "Closed" if registrations >= capacity else "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


registration_system(100, 105)
```
